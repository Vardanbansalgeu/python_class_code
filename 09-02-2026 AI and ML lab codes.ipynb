```python
import pandas as pd
```

```python
df=pd.read_csv("Students1.csv")
```

```python
print(df)
```

```python
df.columns = df.columns.str.strip()
cols_to_normalize = ['Height', 'Weight', 'Stipend']
for col in cols_to_normalize:
    min_val = df[col].min()
    max_val = df[col].max()
    df[col] = (df[col] - min_val) / (max_val - min_val)
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])
```

```python
print("\nStandardized Data:\n", df_scaled)
```
