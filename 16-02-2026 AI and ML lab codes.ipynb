```python
visited = set()
path = []
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E", "F"],
    "C": ["A", "D", "F"],
    "D": ["B", "C", "E", "F"],
    "E": ["B", "D", "F"],
    "F": ["C", "D", "E", "B"]
}
def dfs(node, goal):
    if node == goal:
        path.append(node)
        return True
    if node not in visited:
        visited.add(node)
        path.append(node)
        for neighbor in graph[node]:
            if dfs(neighbor, goal):
                return True
        path.pop()
    return False
start = input("Enter start node: ").strip().upper()
goal = input("Enter goal node: ").strip().upper()

if dfs(start, goal):
    print("DFS Path:", " -> ".join(path))
else:
    print("No path found")
```

```python
from collections import deque
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E", "F"],
    "C": ["A", "D", "F"],
    "D": ["B", "C", "E", "F"],
    "E": ["B", "D", "F"],
    "F": ["C", "D", "E", "B"]
}
def bfs(start, goal):
    visited = set()
    queue = deque([[start]])
    order = []
    while queue:
        path = queue.popleft()
        node = path[-1]
        if node not in visited:
            visited.add(node)
            order.append(node)
            if node == goal:
                return order, path
            for neighbor in graph[node]:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
    return order, None

start = input("Enter start node: ").strip().upper()
goal = input("Enter goal node: ").strip().upper()

order, path = bfs(start, goal)
print("BFS Traversal Order:", " -> ".join(order))
if path:
    print("BFS Path:", " -> ".join(path))
else:
    print("No path found")
```
